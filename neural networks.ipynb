```python
import matplotlib.pylab as plt
import numpy as np
```

```python
# sigmoid fn-act. fn
x=np.arange(-8,8,0.1)
f=1/(1+np.exp(-x))
plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
#flexibility due to bias in adjusting weights
#without bias
w1=0.5
w2=1.0
w3=2.0
l1='w=0.5'
l2='w=1.0'
l3='w=2.0'
for weight,label_name in[(w1,l1),(w2,l2),(w3,l3)]:
    f=1/(1+np.exp(-x*weight))
    plt.plot(x,f,label=label_name)
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.show()
```

```python
#with bias
w = 5.0
b1 = -8.0
b2 = 0.0
b3 = 8.0
b4 = 16.0
l1 = 'b = -8.0'
l2 = 'b = 0.0'
l3 = 'b = 8.0'
l4 = 'b = 16.0'
for b, l in [(b1, l1), (b2, l2), (b3, l3),(b4,l4)]:
    f = 1 / (1 + np.exp(-(x*w+b)))
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_wb(x)')
plt.legend(loc=2)
plt.show()
```
